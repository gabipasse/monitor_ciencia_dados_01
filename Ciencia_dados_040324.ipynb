**.describe() retorna medidas estatísticas sobre as colunas de um DataFrame, como média, percentis, desvio-padrão**

```python
# retornando as medidas estatísticas para todas as colunas do dataframe cidades
cidades.describe()
```

O motivo de o método describe() quando utilizado no dataframe cidades só retornar 2 colunas e não as 5 (totais): Essas são as únicas colunas desse dataframe compostas totalmente por valores númericos. Não é possível cácular mínimo ou média para objetos do tipo string, etc

1.  "count" retorna a quantidade de registros para cada coluna (númerica)

2.  "mean" retorna o valor de média para os valores agrupados por coluna (númerica)

3.  "std" retorna o desvio padrão para os valores agrupados por coluna (númerica)

4.  "min" retorna o valor mínimo para os valores agrupados por coluna (númerica)

5.  "25%" retorna o valor que define a separação entre o primeiro quartil e o segundo quartil para cada coluna (númerica)

6.  "50%" retorna o valor que define a separação entre o segundo quartil e o terceiro quartil para cada coluna (númerica)

7.  "75%" retorna o valor que define a separação entre o terceiro quartil e o quarto quartil para cada coluna (númerica)

8.  "max" retorna o valor máximo para os valores agrupados por coluna (númerica)

```python
# retornando as medidas estatísticas apenas para a coluna "Habitantes" do dataframe cidades
cidades['Habitantes'].describe()
```

```python
# retornando as medidas estatísticas para colunas específicas do dataframe cidades
cidades[['Habitantes', 'Salário-Médio']].describe()
```

**.astype() Utilizado para converter o tipo dos dados presentes em determinada(s) coluna(s) do dataframe original**

```python
# tipo de dados presentes na coluna "Habitantes" nao foi modificado
cidades['Habitantes'].dtypes
```

atenção que, em determinados casos, simplesmente chamar o método não significa modificar o dataframe original. Nesses casos, para modificar a partir da função, a coluna que se deseja modificar deve ser assigned A PARTIR da chamada de método

essa mesma lógica funciona para todos os demais métodos de modificação disponibilizados pelo pandas

```python
# assignando a coluna "Habitantes" do tipo original (int) para object
cidades['Habitantes'] = cidades['Habitantes'].astype('object')
```

```python
# agora sim. 'O' é o mesmo que 'object'
cidades['Habitantes'].dtypes
```

```python
# modificando o tipo dos dados presentes na coluna Habitantes de volta pra int64
cidades['Habitantes'] = cidades['Habitantes'].astype('int64')
```

```python
# :D
cidades['Habitantes'].dtypes
```

atenção pois a conversão entre tipos não é irrestrita. Converter de object para int64 pode dar erro caso a coluna original possua, por exemplo, string com caracteres não númericos

```python
try:

  cidades['Estado'] = cidades['Estado'].astype('int64')

except ValueError as e:

  print("Erro ao converter os dados para int64:", e)
```

**.loc Permite tanto fazer a seleção utilizando diretamente os rótulos do DataFrame, ou seja, os nomes atribuídos às linhas e colunas qaunto utilizando os indexes seja das linhas ou colunas**

no formato `.loc[a:b , 'nome_coluna']`:

1. a é o primeiro index no slicing das linhas (registros)

2. b é o último index no slicing das linhas (registros)

3. Ou seja, quero da linha com index a até a linha com index b

4. 'nome_coluna' é o nome da coluna ou colunas que deseja retornar no slicing

```python
# como não utilizou "," dentro do loc, retorna apenas refentes às colunas
cidades.loc[0:2]
```

```python
# retornando as linhas (registros) que vão do index 0 até o 2, mas apenas da coluna "Cidade"
cidades.loc[0:2, 'Cidade']
```

```python
# retornando as linhas (registros) que vão do index 0 até o 2, mas apenas das colunas "Cidade" e "Estado"
cidades.loc[0:2, ['Cidade', 'Estado']]
```

para retornar apenas uma linha, utilize `.loc[a]`

```python
# retornando apenas a linha de index 2, apenas a coluna "Cidade"
cidades.loc[2, 'Cidade']
```

```python
# retornando a linha de index 2 com todas as colunas
cidades.loc[2]
```

para retornar todas as linhas, utilize `.loc[:]`

```python
cidades.loc[:]
```

no formato `.iloc[a:b , c:d]`:

1. a é o primeiro index no slicing das linhas (registros)

2. b é o último index no slicing das linhas (registros)

3. Ou seja, quero da linha com index a até a linha com index b - 1

4. c é o primeiro index no slicing das colunas

5. d é o último index no slicing das colunas

6. Ou seja, quero da coluna com index c até a coluna com index d - 1

para retornar todas as linhas, utilize `.iloc[:, a]` ou `iloc[:, a:b]`

para retornar todas as colunas, utilize `.iloc[a, :]` ou `iloc[a:b, :]`

```python
# retornando todas as colunas mas apenas a linha (registro) de index 1
cidades.iloc[1, :]
```

```python
# retornando as linhas (registros) que vão do index 1 até o 2 (3 é exclusivo),
# mas apenas da coluna 1 (2 é exclusivo)
cidades.iloc[1:3, 1:2]
```

para retornar apenas uma linha, utilize `.iloc[a]` ou `iloc[a:a+1]`

```python
# retornando apenas a linha (registro) de index 1
cidades.iloc[1]
```

```python
# retornando apenas a linha (registro) de index 1(2 é exclusivo)
cidades.iloc[1:2]
```

para retornar apenas uma coluna, utilize `.iloc[:, a]` ou `iloc[:, a:a+1]`

```python
# retornando todas as linhas e apenas a coluna de index 1
cidades.iloc[:, 1]
```

```python
# retornando todas as linhas e apenas a coluna de index 1(2 é exclusivo)
cidades.iloc[:, 1:2]
```

**.query() Permite filtrar um DataFrame usando uma expressão**

# **Combinando DataFrames**

```python
cidades_salarios = pd.DataFrame(
[
['Paraná', 'Londrina', 1356.00],
['São Paulo', 'São Carlos', 1508.00],
['Santa Catarina', 'Florianópolis', 1798.00],
['Paraná', 'Curitiba', 2293.00],
['São Paulo', 'Campinas', 1710.00]
], columns=['Estado', 'Cidade', 'Salário-Médio'])
```

```python
cidades_salarios
```

```python
cidades_pop = pd.DataFrame(
[
['Londrina', 575377],
['São Carlos', 254484],
['Florianópolis', 508826],
['Curitiba', 1963726],
['Campinas', 1223237]
], columns=['Cidade', 'Habitantes'])
```

```python
cidades_pop
```

**.merge() Concatena dataframes que possuam pelo menos uma coluna em comum. O argumento how="outer" é utilizado para realizar full joins**

```python
# concatena os valores dos dataframes cidades_salarios e cidades_pop de acordo com igualdades no valore da coluna "Cidade"
pd.merge(left=cidades_salarios, right=cidades_pop, on="Cidade")
```

```python
# realizando um full join entre os dataframes cidades_salarios e cidades_pop
pd.merge(left=cidades_salarios, right=cidades_pop, on="Cidade", how="outer")
```

**.concat() Enquanto na função merge() DataFrames são combinados a partir de valores em colunas compartilhadas, a concatenação possibilita combinar dois DataFrames com as mesmas colunas**

```python
cidades_df1 = pd.DataFrame(
[
['Londrina', 575377],
['São Carlos', 254484],
['Florianópolis', 508826],
], columns=['Cidade', 'Habitantes'])
```

```python
cidades_df1
```

```python
cidades_df2 = pd.DataFrame(
[
['Curitiba', 1963726],
['Campinas', 1223237]
], columns=['Cidade', 'Habitantes'])
```

```python
cidades_df2
```

```python
# concatenando os dataframes cidades_df1 e cidades_df2.
# Ignorando os indexes dos dataframes originais, os definindo no dataframe resultante a partir do 0
pd.concat([cidades_df1, cidades_df2], ignore_index=True)
```
